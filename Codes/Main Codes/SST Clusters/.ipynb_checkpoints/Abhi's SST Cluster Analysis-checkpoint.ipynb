```python
import xarray 
import os
import netCDF4
import numpy as np
import matplotlib.pyplot as plt
import imdlib as imd
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import scipy.io
import matplotlib.image as mpimg
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
cluster4_1 = pd.read_csv(r"C:\Users\ratho\Desktop\Software Group Project - III\Climate Data Modelling\Our Work\Codes\Main Codes\SST Clusters\SSTA 4 Cluster Number 1901-2010.csv",index_col=0)
cluster4_2 = pd.read_csv(r"C:\Users\ratho\Desktop\Software Group Project - III\Climate Data Modelling\Our Work\Codes\Main Codes\SST Clusters\SSTA 4 Cluster Number 2011-2022.csv",index_col=0)
```

```python
cluster_4=pd.concat([cluster4_1,cluster4_2])
cluster_4=cluster_4.set_index(i for i in range(122))
```

```python
cluster_4['Cluster Number'] = cluster_4['Cluster Number'].replace(0,10)
cluster_4['Cluster Number'] = cluster_4['Cluster Number'].replace(1,11)
cluster_4['Cluster Number'] = cluster_4['Cluster Number'].replace(2,12)
cluster_4['Cluster Number'] = cluster_4['Cluster Number'].replace(3,13)

cluster_4['Cluster Number'] = cluster_4['Cluster Number'].replace(12,0)
cluster_4['Cluster Number'] = cluster_4['Cluster Number'].replace(10,1)
cluster_4['Cluster Number'] = cluster_4['Cluster Number'].replace(11,2)
cluster_4['Cluster Number'] = cluster_4['Cluster Number'].replace(13,3)
```

```python
# Define a color map for your clusters
colors = {0: 'red', 1: 'blue', 2: 'green', 3: 'purple', 4: 'orange'}

# Create a scatter plot
plt.scatter(cluster_4.index, cluster_4['Cluster Number'], c=cluster_4['Cluster Number'].map(colors), cmap='viridis', s=50)

# Add labels and title
plt.xlabel('Years')
plt.ylabel('Cluster Number')
plt.title('Scatter Plot of Cluster Numbers')

# Show the color legend
legend_labels = {cluster: color for cluster, color in colors.items() if cluster in cluster_4['Cluster Number'].unique()}
legend_handles = [plt.Line2D([0], [0], marker='o', color='w', label=f'Cluster {cluster}', markerfacecolor=color, markersize=10) for cluster, color in legend_labels.items()]
plt.legend(handles=legend_handles, loc='best')

# Display the plot
plt.show()
```
